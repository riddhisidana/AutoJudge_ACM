# **Data preprocessing**

```python
df["full_text"] = (
    df["title"].fillna("") + " " +
    df["description"].fillna("") + " " +
    df["input_description"].fillna("") + " " +
    df["output_description"].fillna("")
)
```

```python
import re
# df["full_text"] = df["full_text"].str.replace(r"<.*?>", " ", regex=True)
df["clean_text"] = (
    df["full_text"]
    .str.lower()
    .str.replace(r"[^a-z0-9\s]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
)
df.head()
```

```python
df["is_easy"] = (df["problem_class"] == "easy").astype(int)
df["is_medium"] = (df["problem_class"] == "medium").astype(int)
df["is_hard"] = (df["problem_class"] == "hard").astype(int)
```

```python
def extract_numeric_features(text):
    text_lower = text.lower()

    # Algorithm groups
    algo_groups = {
        "dp": [
            "dp", "dynamic programming", "knapsack", "bitmask dp",
            "state", "transition", "memoization", "tabulation"
        ],
        "graph": [
            "graph", "bfs", "dfs", "dijkstra", "bellman ford",
            "floyd warshall", "topological", "shortest path",
            "flow", "max flow", "min cut", "matching",
            "strongly connected", "bridges", "articulation points",
            "tree", "lca"
        ],
        "ds": [
            "segment tree", "fenwick", "binary indexed tree",
            "heap", "priority queue", "stack", "queue",
            "deque", "union find", "disjoint set", "sparse table"
        ],
        "math": [
            "modulo", "prime", "gcd", "lcm", "combinatorics",
            "permutations", "probability", "matrix exponentiation",
            "fft", "fast fourier transform", "number theory"
        ],
        "geometry": [
            "geometry", "convex hull", "sweep line",
            "cross product", "dot product", "orientation"
        ],
        "string": [
            "string", "substring", "palindrome",
            "kmp", "z algorithm", "suffix array",
            "trie", "rolling hash"
        ],
        "greedy": [
            "greedy", "two pointers", "sliding window",
            "interval", "activity selection"
        ]
    }

    # Count per algorithm group
    group_counts = {}
    for group, keywords in algo_groups.items():
        count = sum(text_lower.count(k) for k in keywords)
        group_counts[f"{group}_count"] = np.log1p(count)

    # Math symbols
    math_symbols = "+-*/^=<>(){}[]|&!%"
    math_symbol_count = sum(text.count(sym) for sym in math_symbols)

    # Text length
    text_len = len(text)

    # Constraint-awareness features

    has_constraints = int("≤" in text or "<=" in text_lower or "constraints" in text_lower)

    has_big_n = int("10^5" in text or "10^6" in text or "10^7" in text or "10^" in text)

    has_time_limit = int("time limit" in text_lower or "seconds" in text_lower)

    return {
        "text_length": np.log1p(text_len),
        "math_symbol_count": np.log1p(math_symbol_count),
        "has_constraints": has_constraints,
        "has_big_n": has_big_n,
        "has_time_limit": has_time_limit,
        **group_counts
    }
```

```python
numeric_features_df = df["clean_text"].apply(extract_numeric_features)
numeric_features_df = pd.DataFrame(numeric_features_df.tolist())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=30000,
    ngram_range=(1, 2),
    stop_words="english",
    min_df=5,
    max_df=0.9,
    sublinear_tf=True
)

X_tfidf = tfidf.fit_transform(df["clean_text"])
```

# **Modelling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(numeric_features_df)
```

```python
from scipy.sparse import hstack

X = hstack([X_tfidf, X_num_scaled])
```

```python
import numpy as np
y_class_single = np.argmax(
    df[["is_easy", "is_medium", "is_hard"]].values,
    axis=1
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_class_train, y_class_test, y_score_train, y_score_test = train_test_split(
    X,
    y_class_single,
    df["problem_score"],
    test_size=0.2,
    random_state=42,
    stratify=y_class_single
)
```

**Classification Models**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report
```

```python
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
lr = LogisticRegression(
    solver="lbfgs",
    max_iter=3000,
    random_state=42
)
```

```python
param_grid = {
    "C": [0.01, 0.1, 1, 3, 10],
    "class_weight": [None, "balanced"]
}
grid = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train, y_class_train)

```

```python
best_log_clf = grid.best_estimator_
print("Best parameters:", grid.best_params_)
from sklearn.metrics import accuracy_score, classification_report

y_pred = best_log_clf.predict(X_test)

print("Accuracy:", accuracy_score(y_class_test, y_pred))
print(classification_report(
    y_class_test,
    y_pred,
    target_names=["easy", "medium", "hard"]
))

```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

svm = LinearSVC(
    class_weight="balanced",
    max_iter=5000,
    random_state=42
)

param_grid_svm = {
    "C": [0.01, 0.1, 1, 3, 10]
}

svm_grid = GridSearchCV(
    estimator=svm,
    param_grid=param_grid_svm,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

svm_grid.fit(X_train, y_class_train)

best_svm = svm_grid.best_estimator_
print("Best SVM params:", svm_grid.best_params_)
```

```python
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Tuned Logistic Regression": best_log_clf,
    "Tuned Linear SVM": best_svm
}

for name, model in models.items():
    print("\n" + "="*60)
    print(name)

    y_pred = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_class_test, y_pred))
    print(classification_report(
        y_class_test,
        y_pred,
        target_names=["easy", "medium", "hard"]
    ))
```

```python
def to_binary(y):
    return (y == 2).astype(int)

print("\nHARD vs NOT-HARD COMPARISON\n")

for name, model in models.items():
    print("="*50)
    print(name)

    y_pred = model.predict(X_test)

    print(classification_report(
        to_binary(y_class_test),
        to_binary(y_pred),
        target_names=["Not-Hard", "Hard"]
    ))
```

**Regression Model**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_score_train)

y_pred_lr = lin_reg.predict(X_test)

mae_lr = mean_absolute_error(y_score_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_score_test, y_pred_lr))

print("LINEAR REGRESSION")
print("MAE:", mae_lr)
print("RMSE:", rmse_lr)
```

```python
reg = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

reg.fit(X_train, y_score_train)
y_score_pred = reg.predict(X_test)
mae_rf = mean_absolute_error(y_score_test, y_score_pred)
rmse_rf = np.sqrt(mean_squared_error(y_score_test, y_score_pred))

print("MAE:", mae_rf)
print("RMSE:", rmse_rf)
```

```python
param_dist_gb = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.05, 0.1],
    "max_depth": [2, 3, 4],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "subsample": [ 0.8, 0.9, 1.0]
}
gbr = GradientBoostingRegressor(
    random_state=42
)
gb_random = RandomizedSearchCV(
    estimator=gbr,
    param_distributions=param_dist_gb,
    n_iter=20,
    scoring="neg_root_mean_squared_error",
    cv=5,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

gb_random.fit(X_train, y_score_train)
best_gb = gb_random.best_estimator_
print("Best GB params:", gb_random.best_params_)

```

```python
y_pred_gb = best_gb.predict(X_test)

mae_gb = mean_absolute_error(y_score_test, y_pred_gb)
rmse_gb = np.sqrt(mean_squared_error(y_score_test, y_pred_gb))

print("GRADIENT BOOSTING REGRESSION")
print("MAE:", mae_gb)
print("RMSE:", rmse_gb)
```

```python
print("\nREGRESSION MODEL COMPARISON\n")
print(f"Linear Regression → MAE: {mae_lr:.3f}, RMSE: {rmse_lr:.3f}")
print(f"Gradient Boosting → MAE: {mae_gb:.3f}, RMSE: {rmse_gb:.3f}")
print(f"Random Forest → MAE: {mae_rf:.3f}, RMSE: {rmse_rf:.3f}")
```
