# **Data preprocessing**

```python
df["full_text"] = (
    df["title"].fillna("") + " " +
    df["description"].fillna("") + " " +
    df["input_description"].fillna("") + " " +
    df["output_description"].fillna("")
)
```

```python
import re
df["full_text"] = df["full_text"].str.replace(r"<.*?>", " ", regex=True)
df["clean_text"] = (
    df["full_text"]
    .str.lower()
    .str.replace(r"[^a-z0-9\s]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
)
df.head()
```

```python
df["has_big_n"] = df["full_text"].str.contains(r"10\^5|10\^6|10\^7").astype(int)
df["mentions_mod"] = df["full_text"].str.contains(r"modulo|mod").astype(int)
df["mentions_dp_phrase"] = df["full_text"].str.contains("dynamic programming").astype(int)
```

```python
df["text_length"] = df["clean_text"].apply(len)
df["num_math_symbols"] = df["full_text"].str.count(r"[=<>+\-*/]")
```

```python
keywords = ["dp", "graph", "tree", "recursion", "binary search"]
for kw in keywords:
    df[f"kw_{kw}"] = df["clean_text"].str.count(kw)
```

```python
df["is_easy"] = (df["problem_class"] == "easy").astype(int)
df["is_medium"] = (df["problem_class"] == "medium").astype(int)
df["is_hard"] = (df["problem_class"] == "hard").astype(int)
```

```python
def hierarchical_predict(X_input):
    hard_preds = hard_clf.predict(X_input)
    final_preds = np.zeros(len(hard_preds), dtype=int)

    for i in range(len(hard_preds)):
        if hard_preds[i] == 1:
            final_preds[i] = 2  # Hard
        else:
            em_pred = em_clf.predict(X_input[i])
            final_preds[i] = em_pred[0]  # 0=Easy, 1=Medium

    return final_preds
```

```python
def predict_difficulty_score(text):
    # text preprocessing
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)

    # TF-IDF
    text_vec = tfidf.transform([text])

    # numerical features
    extra = [
        len(text),
        text.count("=") + text.count("<") + text.count(">"),
        text.count("dp"),
        text.count("graph"),
        text.count("tree"),
        text.count("recursion"),
        text.count("binary search")
    ]

    extra_scaled = scaler.transform([extra])

    final_vec = hstack([text_vec, extra_scaled])

    return round(reg.predict(final_vec)[0], 2)
```

```python
import streamlit as st
hard_clf = joblib.load("classifier_stage1.pkl")   # Hard vs Not-Hard
em_clf = joblib.load("classifier_stage2.pkl")     # Easy vs Medium
reg = joblib.load("regressor.pkl")                # Random Forest Regressor
tfidf = joblib.load("tfidf.pkl")
scaler = joblib.load("scaler.pkl")

def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text

def build_features(text):
    # TF-IDF features
    text_vec = tfidf.transform([text])

    # Hand-crafted numeric features
    numeric_features = [
        len(text),
        text.count("=") + text.count("<") + text.count(">"),
        text.count("dp"),
        text.count("graph"),
        text.count("tree"),
        text.count("recursion"),
        text.count("binary search")
    ]

    numeric_scaled = scaler.transform([numeric_features])

    # Combine features
    return hstack([text_vec, numeric_scaled])

def predict_class(X):
    """
    Returns:
    0 -> Easy
    1 -> Medium
    2 -> Hard
    """
    hard_pred = hard_clf.predict(X)[0]

    if hard_pred == 1:
        return 2  # Hard
    else:
        em_pred = em_clf.predict(X)[0]
        return em_pred  # 0 = Easy, 1 = Medium

# Streamlit UI
st.set_page_config(page_title="AutoJudge", layout="centered")

st.title("AutoJudge: Programming Problem Difficulty Predictor")

st.markdown("""
Paste the programming problem details below.
The system will predict:
- **Difficulty Class** (Easy / Medium / Hard)
- **Difficulty Score** (Numerical)
""")

# Input Fields (MANDATORY: 3 boxes)
problem_desc = st.text_area("Problem Description", height=150)
input_desc = st.text_area("Input Description", height=100)
output_desc = st.text_area("Output Description", height=100)

# Predict Button
if st.button("Predict Difficulty"):
    if problem_desc.strip() == "":
        st.warning("Please enter the problem description.")
    else:

        combined_text = problem_desc + " " + input_desc + " " + output_desc
        cleaned_text = clean_text(combined_text)


        X_input = build_features(cleaned_text)

        # Predictions
        class_id = predict_class(X_input)
        score_pred = reg.predict(X_input)[0]

        label_map = {0: "Easy", 1: "Medium", 2: "Hard"}

        # results
        st.success(f"Predicted Difficulty Class: **{label_map[class_id]}**")
        st.success(f"Predicted Difficulty Score: **{round(score_pred, 2)}**")
```
